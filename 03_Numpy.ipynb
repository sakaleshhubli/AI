#copy and view

```python
np41 = np.arange(10)
np41
```

```python
np42 = np41.view()
np42
```

```python
print(f"Orignal np41: {np41}")
print(f"View np42: {np42}")
```

```python
np41[0] = 100
print(f"changed np41: {np41}")
print(f"View np42: {np42}")
```

```python
#create a copy
np43 = np41.copy()
np43
```

```python
print(f"Orignal np41: {np41}")
print(f"Copy np43: {np43}")
```

```python
np41[0] = 200
print(f"changed np41: {np41}")
print(f"Copy np43: {np43}")
```

# Iterating

```python
np61 = np.arange(10)
np61
```

```python
for x in np61:
  print(x)
```

```python
np62 = np61.reshape(2, 5)
for x in np62:
  print(x)
```

```python
for x in np62:
  for y in x:
    print(y)
```

```python
# using np.nditer

for x in np.nditer(np62):
  print(x)
```
