Python program to implement the BFS Algo

```python
from collections import deque
```

```python
#define the graph/tree
graph = {
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':[],
    'F':[]
}
```

```python
#perform bfs on a graph strating from given node

def bfs(graph,start_node):
    visited = set()                       #to keep track of visited nodes
    queue = deque([start_node])           #queue to store nodes to visit, starting with root node
    print("BFS Traversal Order: ")
    while queue:                          #while queue is empty
        current_node = queue.popleft()    #dequeue a node from the queue
        if current_node not in visited:   #if the node has not been visited
            print(current_node, end=" ")  #print the node
            visited.add(current_node)     #mark the node as visited
            for neighbor in graph[current_node]:    #for each neighbor of the current node
                if neighbor not in visited:         #if the neighbor has not been visited
                    queue.append(neighbor)          #add the neighbor to the queue
```

```python
#call bfs function
bfs(graph,'A')
```
